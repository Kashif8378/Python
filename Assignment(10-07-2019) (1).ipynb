```python
def reverseFibonacci(n):
    a = [0] * n   
    a[0] = 0
    a[1] = 1
    for i in range(2, n):   
        a[i] = a[i - 2] + a[i - 1]  
    for i in range(n - 1, -1 , -1):   
        print(a[i],end=" ")  
n = 5
reverseFibonacci(n)  
```

```python
def Caps(s):
    spaceCnt=0
    for x in range(len(s)):
        if ord(s[x])==32:
            spaceCnt+=1
        if spaceCnt==1:
            if ord(s[x])>=65 and ord(s[x])<=90:
                print(s[x],end="")
            elif ord(s[x])>=97 and ord(s[x])<=122:
                print(chr(ord(s[x])-32),end="")
        if spaceCnt==2:
            break
    return
Caps("application Testing")
```

```python
#SeriesGeneration1
def SG1(n):
    i=1
    while i<=n:
        print(i,end=" ")
        i=i*3
    return
SG1(100)
```

```python
#SeriesGeneration2
def SG2(n):
    i=1
    while i<=n:
        print(i,end=" ")
        i=i*2
    return
SG2(300)
```

```python
def SG3(n):
    i=1
    t=1
    while i<=n:
        print(t)
        t=i*3+1
        i=i+1
    return
SG3(450)
```
